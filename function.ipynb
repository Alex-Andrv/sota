```python
import torch
from function.function import Function
import matplotlib.pyplot as plt


def draw_3d_function(function: Function, min_v: int = -10, max_v: int = 10):
    x_range = torch.linspace(min_v, max_v, 100)
    y_range = torch.linspace(min_v, max_v, 100)
    X, Y = torch.meshgrid(x_range, y_range)
    concatenated_tensor = torch.cat((X.unsqueeze(0), Y.unsqueeze(0)), dim=0)
    result = function.calculate(concatenated_tensor)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X.numpy(), Y.numpy(), result.numpy(), cmap='viridis')
    # Set labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    # Show the plot
    plt.show()
```

```python
from function.middle.matyas import Matyas

draw_3d_function(Matyas(torch.empty(2, 3)), -100, 100)
```

```python
from function.middle.beale import Beale

draw_3d_function(Beale(torch.empty(2, 3)), -100, 100)
```

```python
from function.easy.vincent import Vincent


draw_3d_function(Vincent(torch.empty(2, 3)), -100, 100)
```
